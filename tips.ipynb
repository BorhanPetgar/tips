```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading Data

```python
tips = sns.load_dataset('tips')
df = tips.copy()
df.head(10)
```

```python
df.describe().T
```

```python
df.corr()
```

# Visualizing

```python
sns.set_style('darkgrid')
sns.relplot(x='total_bill', y='tip', data=df)
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='day')
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', col='day', data=df, col_wrap=2)
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='sex')
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='sex', style='time')
```

```python
sns.catplot(x='day', y='total_bill', data=df, hue='sex')
```
